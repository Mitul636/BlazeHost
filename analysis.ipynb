This is a simple example showing how Python code runs interactively.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create some sample data
data = {
    'Year': [2022, 2023, 2024, 2025, 2026],
    'Users': [100, 250, 500, 800, 1200]
}

df = pd.DataFrame(data)
print("--- Data Preview ---")
print(df)
```

```python
# Create a simple line plot
plt.plot(df['Year'], df['Users'], marker='o', color='blue')
plt.title('Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Users')
plt.grid(True)
plt.show()
```
